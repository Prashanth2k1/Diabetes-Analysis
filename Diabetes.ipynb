```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
desc = df.describe()
print("Descriptive Statistics:\n")
print(desc)
```

```python
mean_vals = df.mean()
print("\nMean:\n", mean_vals)
```

```python
var_vals = df.var()
print("\nVariance:\n", var_vals)
```

```python
skew_vals = df.skew()
print("\nSkewness:\n", skew_vals)
```

```python
kurt_vals = df.kurt()
print("\nKurtosis:\n", kurt_vals)
```

```python
corr = df.corr()

print("\nCorrelation Matrix:\n")
print(corr)

print("\nCorrelation with Outcome:\n")
print(corr["Outcome"].sort_values(ascending=False))
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x="Glucose", y="BMI", hue="Outcome")
plt.title("Relational: Glucose vs BMI by Outcome")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df, x="Outcome", y="Glucose")
plt.title("Categorical: Glucose Distribution by Outcome")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.histplot(df["Glucose"], bins=30, kde=True)
plt.title("Statistical: Glucose Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
